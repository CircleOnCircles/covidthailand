```python
from tableauscraper import TableauScraper as TS

url = "https://public.tableau.com/views/PlayerStats-Top5Leagues20192020/OnePlayerSummary"

ts = TS()
ts.loads(url)
```

```python
workbook = ts.getWorkbook()
```

```python
for i, t in enumerate(workbook.worksheets):
    print(f"{i} worksheet name : {t.name}") #show worksheet name
    print(t.data) #show dataframe for this worksheet
```

```python
ws = ts.getWorksheet("ATT MID CREATIVE COMP")
```

```python
# show selectable values
selections = ws.getSelectableItems()
print(selections)
```

```python
for s in selections:
    print(s['column'])
```

```python
# select that value
dashboard = ws.select("Player Unique", "Wu LeiEspanyol")
```

```python
# display worksheets
for t in dashboard.worksheets:
    print(t.data)
```
